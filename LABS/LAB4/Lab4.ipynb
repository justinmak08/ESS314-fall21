```python
import numpy as np
import matplotlib.pyplot as plt

t_refracted = np.zeros(10)
t_direct = np.zeros(10)
v1 = 2000
v2 = 4000
geophoneDistances = np.linspace(10,100,10)
d = 20
t_refracted = (geophoneDistances/v2) + 2*d*(np.sqrt((v2**2-v1**2))/(v1*v2))
t_direct = geophoneDistances / v1

plt.plot(geophoneDistances, t_refracted, 'g-', label='Refracted Wave Time Curve')
plt.plot(geophoneDistances, t_direct, 'r-', label='Direct Wave Time Curve')
plt.xlabel('Distance From Source to Geophone')
plt.legend(loc='upper right')
plt.ylabel('Wave Travel Time')
```
